```python
from src.libs.lib import *
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
df_original = pd.read_csv("src/data/quantidade_de_motoristas_por_estado.csv")
df_original["Quantidade de Motoristas"] = df_original["Quantidade de Motoristas"].astype(int)
#2.6s
```

```python
df_original
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
df = df_original.copy()
# Aplicar KMeans com 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
df['tier'] = kmeans.fit_predict(df[['Quantidade de Motoristas']])

# Ordenar tiers por média para nomear corretamente
tier_means = df.groupby('tier')['Quantidade de Motoristas'].mean().sort_values()
tier_labels = {tier_means.index[0]: 'Poucos motoristas',
               tier_means.index[1]: 'Alguns motoristas',
               tier_means.index[2]: 'Muitos motoristas'}

df['tier_label'] = df['tier'].map(tier_labels)
df = df.sort_values("tier")
display(df)
```

```python
df.to_csv('src/data/quantidade_de_motoristas_por_estado_resultados.csv')
```
