✅ BASE ANALISADA

- Total de contratos analisados: 18.499

- Contratos com atuação do Falcon: 3.117

- Proporção com Falcon: 16,9%

✅ SCORE DO TRABALHO DO CONTRATO

Índice para mensurar o esforço operacional exigido por cada contrato, considerando três dimensões:

- qt_contrato_alterado: número de vezes que o contrato foi alterado;

- qt_contrato_postergado: quantidade de adiamentos registrados;

- janela_eventos_horas: duração total da janela de eventos do contrato.

Transformações aplicadas: log1p + normalização (MinMaxScaler) + média das variáveis.

✅ RESULTADO:

- Contratos com atuação do Falcon apresentam score de trabalho significativamente maior em todos os tipos analisados.

- Testes estatísticos (t de Student e Kolmogorov-Smirnov) confirmam a diferença (p < 0.05). 

- Visualmente, os contratos com Falcon estão mais concentrados em scores altos.

✅ CONCLUSÃO: 

A atuação do Falcon está associada a contratos com mais alterações, adiamentos e maior janela de eventos, sendo o score desenvolvido se mostrou eficaz para capturar essa complexidade.

```python
df = df_original.copy()
from sklearn.preprocessing import MinMaxScaler

df['alterado_log'] = np.log1p(df['qt_contrato_alterado'])
df['postergado_log'] = np.log1p(df['qt_contrato_postergado'])
df['janela_log'] = np.log1p(df['janela_eventos_horas'])

scaler = MinMaxScaler()
df['alterado_norm'] = scaler.fit_transform(df[['alterado_log']])
df['postergado_norm'] = scaler.fit_transform(df[['postergado_log']])
df['janela_norm'] = scaler.fit_transform(df[['janela_log']])

df['score_trabalho_contrato'] = (df['alterado_norm'] + df['postergado_norm'] + df['janela_norm']) / 3
df[['score_trabalho_contrato']].describe()
```

```python
from scipy.stats import ttest_ind, ks_2samp

# Garantir que 'atuacao_falcon' está como booleano
df['atuacao_falcon_bool'] = df['atuacao_falcon'].astype(str).str.lower().str.strip() == 'sim'

# Lista de tipos
tipos = df['de_type'].dropna().unique()

# Resultados por tipo
resultados_tipo = []

for tipo in tipos:
    df_tipo = df[df['de_type'] == tipo]
    score_com_falcon = df_tipo[df_tipo['atuacao_falcon_bool']]['score_trabalho_contrato']
    score_sem_falcon = df_tipo[~df_tipo['atuacao_falcon_bool']]['score_trabalho_contrato']
    
    media_com = score_com_falcon.mean()
    media_sem = score_sem_falcon.mean()
    p_t = ttest_ind(score_com_falcon, score_sem_falcon, equal_var=False).pvalue
    p_ks = ks_2samp(score_com_falcon, score_sem_falcon).pvalue
    
    resultados_tipo.append({
        'Tipo': tipo,
        'Média com Falcon': media_com,
        'Média sem Falcon': media_sem,
        'Qtd com Falcon': len(score_com_falcon),
        'Qtd sem Falcon': len(score_sem_falcon),
        'Diferença': media_com - media_sem,
        'p-valor t': p_t,
        'p-valor KS': p_ks,
        'Significativo t (5%)': p_t < 0.05,
        'Significativo KS (5%)': p_ks < 0.05
    })

# Gerar DataFrame com os resultados
df_resultados_tipo = pd.DataFrame(resultados_tipo)
df_resultados_tipo.sort_values('Diferença', ascending=False, inplace=True)
df_resultados_tipo.reset_index(drop=True, inplace=True)

# Visualizar
df_resultados_tipo
```

# score_contexto


✅ SCORE DE CONTEXTO

Índice que mede a complexidade operacional de cada combinação de estado (UF) e tipo de contrato (Ajudante, Coleta/Entrega, Diária), a partir de duas variáveis-chave:

Considera a média duas variáveis:

- % de contratos com atuação do Falcon (pct_falcon)

- % de motoristas distintos por contrato (pct_motoristas_distintos)


✅  Classificação do Score

Foram calculados os quartis (Q1 e Q3) do score_contexto por tipo de contrato.

Cada combinação uf x de_type foi classificada em Alta, Média e Baixa, com os seguintes valores de corte (Q1 e Q3):

Tipo	         Q1	     Q3
Ajudante	    0.244	0.510
Coleta/Entrega	0.203	0.285
Diária	        0.285	0.440


🟥 Estados com contexto mais crítico ("Alta") -> indica regiões com maior esforço de suporte.

DF: Alta em todos os tipos

PB: Alta em Ajudante (0.92) e Coleta/Entrega (0.53)

PA, RN, RO: Alta em Diária

AL: Alta em Coleta/Entrega (0.34) e Diária (1.00)

🟩 Estados com contexto mais estável ("Baixa") -> sugere contextos mais previsíveis ou maduros.

MG, MS, SP: predominância de classificações "Baixa"

CE, PE: "Baixa" em Ajudante e Coleta/Entrega


```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from scipy.stats import ttest_rel

# Copiar o novo DataFrame para preservar original
df_linhas = df.copy()

# Binário da atuação do Falcon
df_linhas['atuacao_falcon_bin'] = df_linhas['atuacao_falcon'].astype(str).str.lower().map({'sim': 1, 'nao': 0}).fillna(0).astype(int)

# Calcular contexto por uf + de_type
contexto = df_linhas.groupby(['uf', 'de_type']).agg(
    pct_falcon=('atuacao_falcon_bin', 'mean'),
    diversidade_motoristas=('driver_id', pd.Series.nunique),
    total_contratos=('driver_id', 'count')
).reset_index()
contexto['pct_motoristas_distintos'] = contexto['diversidade_motoristas'] / contexto['total_contratos']

# Normalizar contexto
scaler_contexto = MinMaxScaler()
contexto['pct_falcon_norm'] = scaler_contexto.fit_transform(contexto[['pct_falcon']])
contexto['pct_motoristas_distintos_norm'] = scaler_contexto.fit_transform(contexto[['pct_motoristas_distintos']])
contexto['score_contexto'] = (contexto[['pct_falcon_norm', 'pct_motoristas_distintos_norm']].mean(axis=1))

# eliminar as linhas em que não se tem informações
contexto = contexto[contexto['score_contexto'] > 0]

resumo_contexto = contexto[['uf', 'de_type', 'score_contexto']]
contexto.head()
```

```python
tabela_contexto = contexto.pivot_table(
    index='uf',
    columns='de_type',
    values='score_contexto',
    aggfunc='mean' 
).round(2) 
tabela_contexto
```

```python

# Calcular Q1 e Q3 por tipo
q1_q3_por_tipo = contexto.groupby('de_type')['score_contexto'].quantile([0.25, 0.75]).unstack()
q1_q3_por_tipo.columns = ['Q1', 'Q3']
print("Valores de corte por tipo de contrato (Q1 e Q3):")
print(q1_q3_por_tipo.round(3))

# Juntar os valores de Q1 e Q3 ao dataframe original (apenas uma vez)
contexto_classificado = contexto.merge(q1_q3_por_tipo, on='de_type', how='left')
```

```python
# 
condicoes = [
    contexto_classificado['score_contexto'] <= contexto_classificado['Q1'],
    (contexto_classificado['score_contexto'] > contexto_classificado['Q1']) &
    (contexto_classificado['score_contexto'] <= contexto_classificado['Q3']),
    contexto_classificado['score_contexto'] > contexto_classificado['Q3']
]

categorias = ['Baixa', 'Média', 'Alta']

contexto_classificado['categoria_score_contexto'] = np.select(
    condlist=condicoes,
    choicelist=categorias,
    default='Indefinido'  # evita conflito de tipo
)

# Pivot para criar a tabela de dupla entrada
tabela_categorias = contexto_classificado.pivot_table(
    index='uf',
    columns='de_type',
    values='categoria_score_contexto',
    aggfunc='first'
)

# Função de coloração
def colorir_categoria(val):
    cores = {
        'Baixa': 'background-color: #a1d99b',   # verde claro
        'Média': 'background-color: #fddc6c',   # amarelo claro
        'Alta':  'background-color: #fc9272'    # vermelho claro
    }
    return cores.get(val, '')

# Aplicar estilo visual com .map()
tabela_categorias_contexto = tabela_categorias.style.map(colorir_categoria)


tabela_categorias_contexto
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from scipy.stats import ttest_rel

# Copiar o novo DataFrame para preservar original
df_so_falcon = df_merge.loc[df_merge.atuacao_falcon	 == 'sim'].copy()
# Aplicar log1p
df_so_falcon['qt_atividades_log'] = np.log1p(df_so_falcon['qt_atividades_falcon'])
df_so_falcon['qt_horas_log'] = np.log1p(df_so_falcon['qt_horas_falcon'])

# Normalizar
scaler_falcon = MinMaxScaler()
df_so_falcon['qt_atividades_norm'] = scaler_falcon.fit_transform(df_so_falcon[['qt_atividades_log']])
df_so_falcon['qt_horas_norm'] = scaler_falcon.fit_transform(df_so_falcon[['qt_horas_log']])

# Score de trabalho do Falcon
df_so_falcon['score_trabalho_falcon'] = (
    df_so_falcon['qt_atividades_norm'] + df_so_falcon['qt_horas_norm']
) / 2

df_so_falcon
```

```python
df_so_falcon = df_so_falcon[['mes', 'uf', 'de_type', 'houve_candidatura', 'houve_selecionado',
       'contrato_expirou', 'foi_finalizada', 'foi_cancelada', 'atuacao_falcon',
       'freight_id', 'driver_id', 'janela_eventos_horas',
       'qt_contrato_alterado', 'qt_contrato_postergado',
       'qt_total_eventos_contrato', 'qt_selecionados', 'qt_atividades_falcon',
       'qt_horas_falcon', 'score_trabalho_contrato', 'atuacao_falcon_bool',
       'score_contexto',  'score_trabalho_falcon']]
df_so_falcon
```

```python
# Penalização proporcional
df_so_falcon.loc[:, 'score_trabalho_falcon_contexto'] = (
    df_so_falcon['score_trabalho_falcon'] * (1 - df_so_falcon['score_contexto'])
)

# Normalização
scaler = MinMaxScaler()
df_so_falcon.loc[:, 'score_trabalho_falcon_contexto'] = scaler.fit_transform(
    df_so_falcon[['score_trabalho_falcon_contexto']]
)

df_so_falcon_resumo = df_so_falcon[['mes', 'uf', 'de_type', 'score_trabalho_contrato',	'score_trabalho_falcon', 'score_trabalho_falcon_contexto']]
df_so_falcon_resumo.head()
```

```python
from scipy.stats import friedmanchisquare, ttest_rel
import pandas as pd

resultados = []

for tipo, grupo in df_so_falcon_resumo.groupby('de_type'):
    if len(grupo) >= 10:
        # Friedman para os 3 scores
        stat_friedman, p_friedman = friedmanchisquare(
            grupo['score_trabalho_contrato'],
            grupo['score_trabalho_falcon'],
            grupo['score_trabalho_falcon_contexto']
        )
        
        # Testes post-hoc t pareado
        t1, p1 = ttest_rel(grupo['score_trabalho_contrato'], grupo['score_trabalho_falcon'])
        t2, p2 = ttest_rel(grupo['score_trabalho_contrato'], grupo['score_trabalho_falcon_contexto'])
        t3, p3 = ttest_rel(grupo['score_trabalho_falcon'], grupo['score_trabalho_falcon_contexto'])

        resultados.append({
            'de_type': tipo,
            'n': len(grupo),
            
            'media_contrato': grupo['score_trabalho_contrato'].mean(),
            'media_falcon': grupo['score_trabalho_falcon'].mean(),
            'media_ajustado': grupo['score_trabalho_falcon_contexto'].mean(),

            'p_friedman': p_friedman,
            'friedman_significativo': p_friedman < 0.05,

            'dif_contrato_vs_falcon': grupo['score_trabalho_falcon'].mean() - grupo['score_trabalho_contrato'].mean(),
            'p_contrato_vs_falcon': p1,
            'signif_contrato_vs_falcon': p1 < 0.05,

            'dif_contrato_vs_ajustado': grupo['score_trabalho_falcon_contexto'].mean() - grupo['score_trabalho_contrato'].mean(),
            'p_contrato_vs_ajustado': p2,
            'signif_contrato_vs_ajustado': p2 < 0.05,

            'dif_falcon_vs_ajustado': grupo['score_trabalho_falcon_contexto'].mean() - grupo['score_trabalho_falcon'].mean(),
            'p_falcon_vs_ajustado': p3,
            'signif_falcon_vs_ajustado': p3 < 0.05
        })

df_resultados_posthoc = pd.DataFrame(resultados)


df_resultados_posthoc
```

Todas as diferenças são estatisticamente significativas (p < 0.05 em todos os casos).

O score Falcon ajustado pelo contexto é consistentemente mais alto que o contrato original em Ajudante e Coleta/Entrega.

```python
import numpy as np
import pandas as pd

# Calcular a penalização
df_penalizacao = df_so_falcon_resumo.copy()
df_penalizacao['penalizacao'] = df_penalizacao['score_trabalho_falcon_contexto'] - df_penalizacao['score_trabalho_falcon']

# 1. Agrupar por UF e de_type
df_penalizacao_agg = df_penalizacao.groupby(['uf', 'de_type'])['penalizacao'].mean().reset_index()

# 2. Calcular Q1 e Q3 por tipo
q1_q3_penalizacao = df_penalizacao_agg.groupby('de_type')['penalizacao'].quantile([0.25, 0.75]).unstack()
q1_q3_penalizacao.columns = ['Q1', 'Q3']

# 3. Merge para comparar com Q1/Q3
df_penalizacao_classif = df_penalizacao_agg.merge(q1_q3_penalizacao, on='de_type')

# 4. Classificação com base nos quartis
condicoes = [
    df_penalizacao_classif['penalizacao'] <= df_penalizacao_classif['Q1'],
    (df_penalizacao_classif['penalizacao'] > df_penalizacao_classif['Q1']) &
    (df_penalizacao_classif['penalizacao'] <= df_penalizacao_classif['Q3']),
    df_penalizacao_classif['penalizacao'] > df_penalizacao_classif['Q3']
]

categorias = ['Alta', 'Média', 'Baixa']  # Penalização alta = mais redução

df_penalizacao_classif['categoria_penalizacao'] = np.select(condicoes, categorias, default='Indefinido')

# 5. Pivot para exibição
tabela_categorias_penalizacao = df_penalizacao_classif.pivot_table(
    index='uf',
    columns='de_type',
    values='categoria_penalizacao',
    aggfunc='first'
)

# 6. Coloração
def colorir_penalizacao(val):
    cores = {
        'Alta': 'background-color: #fc9272',
        'Média': 'background-color: #fddc6c',
        'Baixa': 'background-color: #a1d99b'
    }
    return cores.get(val, '')

# 7. Exibir valores de corte
print("Valores de corte por tipo de contrato (Q1 e Q3) - Penalização:")
print(q1_q3_penalizacao.round(3))

# 8. Tabela colorida
tabela_categorias_penalizacao_styled = tabela_categorias_penalizacao.style.map(colorir_penalizacao)

tabela_categorias_penalizacao_styled 
```

# comparação entre as duas formas de cálculo do trabalho


A análise foi motivada pela necessidade de validar o critério atual de divisão de motoristas entre Leonardo e Donegá.

Objetivo: Comparar duas formas distintas de avaliação do trabalho:

score_trabalho_contrato: baseado nas características observadas do contrato (adiamentos, alterações, duração). -> reuniao crescimento

Avalia a dificuldade observada com base nas características do contrato, refletindo a complexidade percebida durante a realização do contrato.

score_trabalho_falcon_contexto: reflete o esforço operacional do Falcon (número de atividades e horas investidas), ajustado pelo contexto regional (UF + tipo de contrato).
                        
Mede o esforço do time Falcon, representa uma visão operacional, voltada ao trabalho necessário para garantir a execução.

Cálculo da Diferença:

Foi calculada a diferença entre o score_trabalho_contrato e o score_trabalho_falcon_contexto.

Essa diferença foi chamada de comparacao e representa o descolamento entre as duas perspectivas de avaliação.




A introdução do score_trabalho_contrato permitiu uma segunda perspectiva, baseada nos eventos concretos e nas características dos contratos (como adiamentos e alterações).

Ao comparar essas duas métricas, conseguimos identificar o quanto essas avaliações convergem ou se distanciam.

```python
from scipy.stats import ttest_rel
from sklearn.preprocessing import MaxAbsScaler

# 1. Diferença entre scores
df_comparacao = df_penalizacao.copy()
df_comparacao['comparacao'] = df_comparacao['score_trabalho_contrato'] - df_comparacao['score_trabalho_falcon_contexto']

# 2. Normalização entre -1 e 1 (preservando sinal)
scaler = MaxAbsScaler()
df_comparacao['comparacao'] = scaler.fit_transform(df_comparacao[['comparacao']])

# 3. Agregar por UF e tipo
df_comparacao_agg = df_comparacao.groupby(['uf', 'de_type']).agg(
    n_contratos=('comparacao', 'count'),
    comparacao=('comparacao', 'mean')
).reset_index()

import numpy as np

# Evita divisão por zero
df_comparacao['pct_diferenca'] = np.where(
    df_comparacao['score_trabalho_falcon_contexto'] > 0,
    (df_comparacao['score_trabalho_contrato'] - df_comparacao['score_trabalho_falcon_contexto']) / df_comparacao['score_trabalho_falcon_contexto'],
    np.nan
)
df_pct_agg = df_comparacao.groupby(['uf', 'de_type'])['pct_diferenca'].mean().reset_index()
df_pct_agg['pct_diferenca'] = (df_pct_agg['pct_diferenca'] * 100).round(1)  # transforma em %


df_comparativo_final0 = pd.merge(
    df_comparacao_agg,
    df_pct_agg,
    on=['uf', 'de_type'],
    how='left'
)

resultados_sig = []

for (uf, tipo), grupo in df_so_falcon_resumo.groupby(['uf', 'de_type']):
    if len(grupo) >= 20:  # evitar pequenos n
        stat, p = ttest_rel(grupo['score_trabalho_falcon_contexto'], grupo['score_trabalho_falcon'])
        resultados_sig.append({
            'uf': uf,
            'de_type': tipo,
            'p_valor': p,
            'significativo': p < 0.05
        })

df_sig = pd.DataFrame(resultados_sig)
df_sig = df_sig.replace({True: 'SIGNIF.', False: 'NS'})

df_comparativo_final = pd.merge(
    df_comparativo_final0,
    df_sig,
    on=['uf', 'de_type'],
    how='left'
)
df_comparativo_final = df_comparativo_final[['uf', 'de_type', 'n_contratos','comparacao', 'significativo', 'pct_diferenca']]
df_comparativo_final = df_comparativo_final.dropna()
df_comparativo_final.head()
```

📌 Para cada combinação de UF e tipo de contrato, foi avaliado:

- n_contratos: quantidade de contratos com atuação do Falcon entre fevereiro e abril de 2025.

- comparacao: média da diferença entre score_trabalho_contrato e score_trabalho_falcon_contexto, ou seja, se o contrato aparenta mais dificuldade com base nas características do contrato (como adiamentos e alterações) em comparação com o esforço operacional do Falcon ajustado por contexto regional.

- significativo: resultado do teste t pareado (ttest_rel) entre os dois scores. Indica se a diferença entre eles é estatisticamente significativa (p < 0.05), ou seja, se não é fruto de variação aleatória.

- pct_diferenca: percentual que indica quanto o score do contrato foi maior ou menor que o score Falcon com contexto, em média. Valores negativos indicam que o score do Falcon foi mais alto.

```python
# Plotar gráfico de barras
plt.figure(figsize=(14, 6))
sns.barplot(data=df_comparativo_final, x='uf', y='pct_diferenca', hue='de_type')
plt.axhline(0, color='black', linestyle='--')
plt.title('Comparação: Score Contrato x Score Falcon Com Diferença Sigificativa')
plt.ylabel('Diferença')
plt.xlabel('UF')
plt.legend(title='Tipo de Contrato')
plt.tight_layout()
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```
