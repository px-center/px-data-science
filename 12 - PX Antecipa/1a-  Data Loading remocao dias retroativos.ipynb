# df_todos_retroativos

```python
df_todos_retroativos = pd.read_csv("src/data/todos_contratos_dias_retroativos.csv")
df_todos_retroativos = df_todos_retroativos.loc[:, ~df_todos_retroativos.columns.str.contains('^Unnamed')] 
df_todos_retroativos = df_todos_retroativos.dropna()
df_todos_retroativos['created_at'] = pd.to_datetime(df_todos_retroativos['created_at'])
```

```python
df_todos_retroativos
```

```python
def classificar_periodo(data):
    dia = data.day
    if dia <= 10:
        return pd.Timestamp(data.year, data.month, 1)
    elif dia <= 20:
        return pd.Timestamp(data.year, data.month, 11)
    else:
        return pd.Timestamp(data.year, data.month, 21)

df_todos_retroativos['periodo'] = df_todos_retroativos['created_at'].apply(classificar_periodo)


df_todos_retroativos = df_todos_retroativos[
    (df_todos_retroativos['created_at'] >= '2025-02-01') &
    (df_todos_retroativos['created_at'] < '2025-04-21')
]
df_todos_retroativos = df_todos_retroativos[['periodo', 'company_id','company_name',  'freight_id', 'driver_id', 'tipo', 'nivel_px','contract_days']]

df_todos_retroativos
```

```python
todos = df_todos_retroativos[['company_id','company_name','freight_id']].drop_duplicates()
```

```python
df_todos_retroativos.to_csv("src/data/df_todos_retroativos.csv")
```

# df_original

```python
df_original = pd.read_csv("src/data/remocao_dias_retroativos.csv")
df_original = df_original.loc[:, ~df_original.columns.str.contains('^Unnamed')] 
df_original = df_original.dropna()
df_original['contract_days_inicial'] = df_original['contract_days_final'] + df_original['dias_reducao'] 
df_original['created_at'] = pd.to_datetime(df_original['created_at'])
df_original.head()
```

```python
def classificar_periodo(data):
    dia = data.day
    if dia <= 10:
        return pd.Timestamp(data.year, data.month, 1)
    elif dia <= 20:
        return pd.Timestamp(data.year, data.month, 11)
    else:
        return pd.Timestamp(data.year, data.month, 21)

df_original['periodo'] = df_original['created_at'].apply(classificar_periodo)


df_original = df_original[
    (df_original['created_at'] >= '2025-02-01') &
    (df_original['created_at'] < '2025-04-21')
]
df_original = df_original[['periodo', 'freight_id', 'driver_id', 'company_id','tipo',  'nivel_px' ,'contract_days_inicial','contract_days_final','dias_reducao',]]
df_original
```

```python
df_original = df_original[['periodo', 'company_id', 'freight_id','tipo', 'nivel_px', 'contract_days_inicial', 'dias_reducao']]
df_original.head()
```

# df_transportadoras

```python
com_reducao = df_original[['company_id', 'freight_id', 'dias_reducao']]
com_reducao['houve_reducao'] = 'sim'
com_reducao
```
