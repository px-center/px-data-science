```python
from pydub import AudioSegment
import speech_recognition as sr
import os

def transcriber_audio(mp3_path):
    """Converte MP3 para WAV e transcreve com Google Speech Recognition."""
    try:
        wav_path = mp3_path.replace(".mp3", ".wav")
        
        # Converter mp3 para wav
        audio = AudioSegment.from_mp3(mp3_path)
        audio.export(wav_path, format="wav")

        r = sr.Recognizer()
        with sr.AudioFile(wav_path) as source:
            audio_data = r.record(source)

        try:
            texto = r.recognize_google(audio_data, language="pt-BR")
            print("Transcri√ß√£o: " + texto)
            return texto
        except sr.UnknownValueError:
            return "N√£o foi poss√≠vel entender o √°udio."
        except sr.RequestError as e:
            return f"Erro na API: {e}"

    except FileNotFoundError:
        return "Arquivo n√£o encontrado."
    except Exception as e:
        return f"Erro: {e}"

# Exemplo de uso
caminho = "src/data/audios/07609489840/2381160897/07609489840_2381160897_audio_resposta_1.wav"
resultado = transcriber_audio(caminho)

print("\n‚úÖ Resultado:")
print(resultado)
```

```python
import os
import subprocess

# Caminho base onde os √°udios est√£o localizados
base_dir = "./src/data"
falhas = []

# Percorre todas as subpastas
for root, _, files in os.walk(base_dir):
    for file in files:
        if file.endswith(".mp3"):
            mp3_path = os.path.join(root, file)
            wav_path = os.path.splitext(mp3_path)[0] + ".wav"

            print(f"üéß Convertendo: {mp3_path}")

            # Comando FFmpeg
            cmd = ["ffmpeg", "-y", "-i", mp3_path, wav_path]

            try:
                result = subprocess.run(cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
                if result.returncode != 0:
                    print(f"‚ùå Falha ao converter {mp3_path}")
                    falhas.append((mp3_path, result.stderr))
            except Exception as e:
                print(f"‚ùå Erro ao processar {mp3_path}: {e}")
                falhas.append((mp3_path, str(e)))

# Resumo
print("\n‚úÖ Convers√£o finalizada.")
if falhas:
    print(f"\nüö® {len(falhas)} arquivos falharam:")
    for caminho, erro in falhas:
        print(f"- {caminho}")
else:
    print("üéâ Todos os arquivos foram convertidos com sucesso.")
```
