```python
from src.libs.lib import *
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
df_original = pd.read_csv("src/data/quantidade_de_motoristas_por_estado.csv")
df_original["Quantidade de Motoristas"] = df_original["Quantidade de Motoristas"].astype(int)
#2.6s
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = df_original.copy()

# Ordenar em ordem decrescente
df_sorted = df.sort_values(by='Quantidade de Motoristas', ascending=False)

# Gerar gráfico de barras
plt.figure(figsize=(12, 8))
plt.bar(df_sorted['Estado'], df_sorted['Quantidade de Motoristas'])
plt.xticks(rotation=90)
plt.xlabel('Estado')
plt.ylabel('Quantidade de Motoristas')
plt.title('Quantidade de Motoristas por Estado (Ordem Decrescente)')
plt.tight_layout()
plt.show()
```
