```python
from src.libs.lib import *
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df_planilha = pd.read_excel("src/data/Satisfação com a PX(1-188).xlsx")
# Limpar espaços em branco nas colunas
df_planilha.columns = df_planilha.columns.str.strip()

# Renomear
df_planilha = df_planilha.rename(columns={
    'Deixe seu telefone que entraremos em contato.': 'telefone'
})

import re

# Remove tudo que não for número
df_planilha['telefone'] = df_planilha['telefone'].astype(str).str.replace(r'\D', '', regex=True)

# Se tiver mais de 11 dígitos, mantém apenas os últimos 11
df_planilha['telefone'] = df_planilha['telefone'].str[-11:]

# Para números com apenas 10 dígitos, insere o 9 após o DDD
mask_10_digitos = df_planilha['telefone'].str.len() == 10
df_planilha.loc[mask_10_digitos, 'telefone'] = (
    df_planilha.loc[mask_10_digitos, 'telefone'].str[:2] + '9' + df_planilha.loc[mask_10_digitos, 'telefone'].str[2:]
)

# Remove os que não têm exatamente 11 dígitos
df_planilha.loc[df_planilha['telefone'].str.len() != 11, 'telefone'] = pd.NA

# Formata para (XX)XXXXX-XXXX
df_planilha['telefone'] = df_planilha['telefone'].str.replace(
    r'^(\d{2})(\d{5})(\d{4})$', r'(\1)\2-\3', regex=True
)


# Verificar
df_planilha.head(5)
```
