```python
from lib import *
import pandas as pd
import matplotlib.pyplot as plt

pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')


#dfs = load_data_bq()
dfs = load_data_db()
df_diaria = dfs["diaria"]
df_coleta_entrega = dfs["coleta_entrega"]
#55s
```

```python
display(df_diaria.head())
display(df_coleta_entrega.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Supondo que df_diaria j√° esteja definido
df = df_diaria.copy()

fig, ax = plt.subplots(figsize=(20, 10))

# Plota as barras com uma cor aprimorada
bars = ax.bar(df['activity_month'].astype(int), df['avg_contracts_per_month'], 
              color='skyblue', edgecolor='black')

# Configura os r√≥tulos dos eixos e t√≠tulo com fontes maiores
ax.set_xlabel('Activity Month', fontsize=14)
ax.set_ylabel('Avg Contracts per Month', fontsize=14)
ax.set_title('Average Contracts per Activity Month\n Di√°ria \n', fontsize=32)

# Define os ticks do eixo x com rota√ß√£o
ax.set_xticks(df['activity_month'].astype(int))
ax.set_xticklabels(df['activity_month'].astype(int), rotation=45, fontsize=12)

# Adiciona linhas de grade no eixo y para facilitar a leitura dos valores
ax.yaxis.grid(True, linestyle='--', alpha=0.7)

# Anota o valor de cada barra
for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{height:.2f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # deslocamento para cima
                textcoords="offset points",
                ha='center', va='bottom', fontsize=12)

plt.tight_layout()
plt.show()
```
